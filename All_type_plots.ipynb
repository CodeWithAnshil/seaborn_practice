## Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
dt=sns.load_dataset("penguins")
dt
sns.barplot(x='species',y='bill_length_mm',data=dt,hue="sex",palette="PiYG",hue_order=["Female","Male"],errorbar=('ci', 100),err_kws={'color': 'b'},saturation=20,capsize=0.2,alpha=0.8)
plt.show()
```

# Pair Plot 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

dt=sns.load_dataset("tips")
dt
```

```python
m={"Male":'o',"Female":'p'}
sns.pairplot(dt,y_vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="cubehelix",markers=m,)
plt.show()
```

```python
m={"Male":'o',"Female":'p'}
sns.pairplot(dt,y_vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="cubehelix",markers=m,diag_kind='hist')
plt.show()
```

```python
m={"Male":'o',"Female":'p'}
sns.pairplot(dt,x_vars=["total_bill","tip"],y_vars=["size","sex"],hue="sex",palette="cubehelix",markers=m)
plt.show()
```

# (Multiple Plot) Facet_Grid in Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
vars=sns.load_dataset("tips")
fg=sns.FacetGrid(vars,col="sex",hue="day",palette="rocket")
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```
