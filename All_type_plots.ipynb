# Histogram plt

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset("planets")
df
sns.displot(df["mass"],kde=True,rug=True,color="green",log_scale=True)
plt.show()
```

## Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
dt=sns.load_dataset("penguins")
dt
sns.barplot(x='species',y='bill_length_mm',data=dt,hue="sex",palette="PiYG",hue_order=["Female","Male"],errorbar=('ci', 100),err_kws={'color': 'b'},saturation=20,capsize=0.2,alpha=0.8)
plt.show()
```

```python
# sns.set(style="darkgrid")
sns.barplot(x='species',y='bill_length_mm',data=dt,hue="sex",palette="PiYG",alpha=0.8,dodge=False,err_kws={'color': 'b'})
plt.show()
```

# Heatmap Plot


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

randm=np.random.randint(1,10,20).reshape(4,5)
randm
```

```python
sns.heatmap(randm,annot=True,cmap='PiYG',vmin=0,vmax=14,annot_kws={'fontsize':15,'color':'black'},linewidth=10,linecolor='y',)
plt.show()
```
